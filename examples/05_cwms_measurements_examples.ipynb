# cwms-python streamflow measurement endpoint examples
The extent endpoint is a quick way to see what locations have measurements stored and the start and end dates of the measurements


```python
# let's see how many sites have measurements in the national db
extents = cwms.get_measurements_extents()
df = extents.df.copy()
df
```

```python
# check which offices have any measurements stored
df['id.office-id'].unique()
```

```python
# which offices have the most sites w/ measurements
office_counts = df.groupby('id.office-id').size()
office_counts
```

```python
# checkout MVK's measurements
extents = cwms.get_measurements_extents(office_mask='MVK')
df = extents.df.copy()
df
```

```python
# Convert the time columns to datetime objects
df['time-extents.earliest-time'] = pd.to_datetime(df['time-extents.earliest-time'], format='mixed')
df['time-extents.latest-time'] = pd.to_datetime(df['time-extents.latest-time'], format='mixed')
```

```python
# Find the site with the oldest measurement
oldest_site = df.loc[df['time-extents.earliest-time'].idxmin(), 'id.name']
oldest_date = df['time-extents.earliest-time'].min()  # Get the actual date
# Find the site with the newest measurement
newest_site = df.loc[df['time-extents.latest-time'].idxmax(), 'id.name']
newest_date = df['time-extents.latest-time'].max()  # Get the actual date

print(f"Site with the oldest measurement: {oldest_site} (Date: {oldest_date})")
print(f"Site with the newest measurement: {newest_site} (Date: {newest_date})")
```

Let's look at the actual measurement data

```python
# get the measurements
data = cwms.get_measurements(location_id_mask='Vicksburg', office_id='MVK')
data.json
```

```python
# convert to a dataframe
df = data.df.copy()
df
```

```python
# wow almost 9k measurements!
# check for duplicates
len(df['instant'].unique())
```

```python

# ok no duplicates
# let's plot it
import matplotlib.pyplot as plt

df.plot(x='streamflow-measurement.flow', y='streamflow-measurement.gage-height', kind='scatter')
plt.ylabel("Gage Height")
plt.xlabel("Flow")
plt.title("Gage Height vs. Flow")
plt.show()
```

```python
# let's add the rating to the plot
ratingId = 'Vicksburg.Stage;Flow.Linear.Distributed'
rating = cwms.get_current_rating(rating_id=ratingId, office_id='MVK')
df_rating = rating.df.copy()


ax = df.plot(x='streamflow-measurement.flow', y='streamflow-measurement.gage-height', kind='scatter')
df_rating.plot(x='dep', y='ind', ax=ax, kind='line', label=ratingId, color='red') 

plt.ylabel("Gage Height")
plt.xlabel("Flow")
plt.title("Gage Height vs. Flow")
plt.show()
```

```python
# let's just see recent measurements since 2015
# Convert the time columns to datetime objects
df['instant'] = pd.to_datetime(df['instant'], format='mixed')
df_filtered = df[df['instant']>'2015']
df_filtered
```

```python
ax = df_filtered.plot(x='streamflow-measurement.flow', y='streamflow-measurement.gage-height', kind='scatter')
df_rating.plot(x='dep', y='ind', ax=ax, kind='line', label=ratingId, color='red') 

plt.ylabel("Gage Height")
plt.xlabel("Flow")
plt.title("Gage Height vs. Flow - Measurements Since 2015")
plt.show()
```

# Storing Ratings
To store ratings, you must give it a list of dictionaries which contain at a minimum the following:
1. Office ID
2. Name (Location)
3. Number
4. Instant (Time)
5. Gage height or flow
6. Used
7. Height and flow units

```python
# lets set up an CDA instance where we can write
from dotenv import load_dotenv
import os
# grab API variables from .env file
load_dotenv()
APIROOT = os.getenv("API_ROOT")
OFFICE = os.getenv("OFFICE")
APIKEY = os.getenv('API_KEY')
# connect to T7
apiKey = "apikey " + APIKEY
api = cwms.api.init_session(api_root=APIROOT, api_key=apiKey)
```

```python
# for more fields please look at the swagger docs or the get_usgs_measurements.py in cwms data acquisition
example_jsonList = [
    {
        "height-unit": "ft",
        "flow-unit": "cfs",
        "used": False,
        "agency": "USACE",
        "party": "Clark",
        "wm-comments": "This is a test",
        "instant": "2025-04-14T00:00:00Z",
        "id": {"office-id": OFFICE, "name": "Test"},
        "number": "1",
        "streamflow-measurement": {"gage-height": 1, "flow": 1000, "quality": "Poor"},
    }
]
```

```python
cwms.store_measurements(data=example_jsonList, fail_if_exists=True)
```

```python
data = cwms.get_measurements(location_id_mask='Test', office_id=OFFICE)
data.df
```
