# CWMS User Role Examples

Simple examples for user role management with `cwms-python`.

This notebook uses only exposed `cwms` methods:
- `cwms.init_session(...)`
- `cwms.store_user(...)`
- `cwms.update_user(...)`

```python
import os

import cwms
```

```python
# Required environment variables:
# CDA_API_ROOT: e.g. http://localhost:8082/cwms-data/
# CDA_API_KEY: include the 'apikey ' prefix
API_ROOT = os.environ["CDA_API_ROOT"]
API_KEY = os.environ["CDA_API_KEY"]

USER_NAME = "SOME_USER" # Your UPASS ID on T7 / CWBI Test, would be your EDIPI "username" on CWBI
OFFICE_ID = "SWT"

# The role to add temporarily
ROLE_TO_ADD = "CWMS User"

# The full role set to restore after the example
# Replace this with the user's real baseline roles for the office
ORIGINAL_ROLES = ["CWMS User"]

cwms.init_session(api_root=API_ROOT, api_key=API_KEY)
```

## 1) Add A Role

`cwms.store_user` adds role(s) to the target user for an office.

```python
cwms.store_user(
    user_name=USER_NAME,
    office_id=OFFICE_ID,
    roles=[ROLE_TO_ADD],
)

print(f"Added role '{ROLE_TO_ADD}' for user '{USER_NAME}' in office '{OFFICE_ID}'.")
```

## 2) Restore / Replace Roles

`cwms.update_user` replaces the user's roles for that office with the list you pass.

```python
cwms.update_user(
    user_name=USER_NAME,
    office_id=OFFICE_ID,
    roles=ORIGINAL_ROLES,
)

print(f"Restored roles for user '{USER_NAME}' in office '{OFFICE_ID}'.")
```
